### **d - ii) Resolving rowspan and colspan challenges**

Due to the data in this project being taken from Wikipedia, the 'rowspan' and 'colspan' details within this table need to be 'unmerged' when converting to a pandas dataframe, otherwise the data moves to incorrect positions.

```python
data = []
rows = selected_table.find_all('tr')

# Track rowspans
rowspan_dict = {}

for row_idx, row in enumerate(rows):
  cols = row.find_all(['td', 'th'])
  col_idx = 0
  row_data = []
  while col_idx < len(cols):
    col = cols[col_idx]
    colspan = int(col.get('colspan', 1))
    rowspan = int(col.get('rowspan', 1))

    # Add previous rowspan data if any
    if row_idx in rowspan_dict:
      row_data.extend(rowspan_dict[row_idx])
      del rowspan_dict[row_idx]

    # Add the current cell data
    cell_data = col.get_text(strip=True)
    row_data.append(cell_data)

    # Handle colspan
    if colspan > 1:
      for _ in range(1, colspan):
        row_data.append(cell_data)

    # Handle rowspan
    if rowspan > 1:
      for r in range(1, rowspan):
        if row_idx + r not in rowspan_dict:
          rowspan_dict[row_idx + r] = []
        rowspan_dict[row_idx + r].append(cell_data)

    col_idx += colspan

  # Add remaining rowspan data if any
  if row_idx in rowspan_dict:
    row_data.extend(rowspan_dict[row_idx])
    del rowspan_dict[row_idx]

  data.append(row_data)

#Create a pandas DataFrame
df = pd.DataFrame(data)
```

**Death Toll** - This column needs to be split up into lowest estimated death toll and highest estimated death toll, while also removing unnecessary characters, and the new columns can then be converted to integer values.

```python
#dealing with multiple statistics within 'death toll' to create two new columns, lowest and highest death toll

dfCopy['Death toll'] = dfCopy['Death toll'].astype(str)

# Function to extract lower values
def extract_values_low(range):
  # Split on dash and return first value
  if '–' in range:
    lower_value = range.split('–')[0]
  elif '-' in range:
    lower_value = range.split('-')[0]
  else:
    lower_value = range.split('—')[0]

  return lower_value

# Function to extract upper values
def extract_values_high(range):
  # Split on dash and reutrn second value
  if '–' in range:
    upper_value = range.split('–')[1]
  elif '-' in range:
    upper_value = range.split('-')[1]
  elif '—' in range:
    upper_value = range.split('—')[1]
  else:
    upper_value = range
  return upper_value

# Initialize lists to store the extracted values
lowestDeathToll = []
highestDeathToll = []

countRow = len(dfCopy)

for i in range(countRow):
  deathTollCheck = dfCopy['Death toll'][i]
  #if dashed within 'Death Toll', split then assign to two lists
  if '–' or '-' or '—' in deathTollCheck:
    lowest = extract_values_low(deathTollCheck)
    highest = extract_values_high(deathTollCheck)
    # Append to the lists
    lowestDeathToll.append(lowest)
    highestDeathToll.append(highest)
  else:
  #else assign to both lists whith the same value
    lowestDeathToll.append(deathTollCheck)
    highestDeathToll.append(deathTollCheck)

# Add the new columns to the DataFrame
dfCopy['Lowest Estimated Death Toll'] = lowestDeathToll
dfCopy['Highest Estimated Death Toll'] = highestDeathToll

```

```python
#removing additional information from lowest and highest estimated death toll column in order to change the column into a set of integer dtypes

dfCopy['Lowest Estimated Death Toll'] = dfCopy['Lowest Estimated Death Toll'].str.rstrip('+')
dfCopy['Highest Estimated Death Toll'] = dfCopy['Highest Estimated Death Toll'].str.rstrip('+')

dfCopy['Lowest Estimated Death Toll'] = dfCopy['Lowest Estimated Death Toll'].str.replace(',', '')
dfCopy['Highest Estimated Death Toll'] = dfCopy['Highest Estimated Death Toll'].str.replace(',', '')

dfCopy['Lowest Estimated Death Toll'] = dfCopy['Lowest Estimated Death Toll'].str.replace('[13]', '')
dfCopy['Highest Estimated Death Toll'] = dfCopy['Highest Estimated Death Toll'].str.replace('[13]', '')

dfCopy['Lowest Estimated Death Toll'] = pd.to_numeric(dfCopy['Lowest Estimated Death Toll'], errors='coerce')
dfCopy['Highest Estimated Death Toll'] = pd.to_numeric(dfCopy['Highest Estimated Death Toll'], errors='coerce')


dt_columns_to_subtract = ['Highest Estimated Death Toll', 'Lowest Estimated Death Toll']

# To create range of estimated death tolls, we subtract the lowest from highest and create a new colum
dfCopy['Death Toll Range'] = df[dt_columns_to_subtract[0]] - df[dt_columns_to_subtract[1]]
```

**Date** - This column needs to be split up into a start date and end date and converted into a datetime datatype, in order to work out the amount of days each event took to then be analysed.

```python
df['Date'] = df['Date'].astype(str)

# Initialize lists to store the extracted values
lowestDate = []
lowestDay = []
lowestMonth = []
highestDate = []
highestDay = []
highestMonth = []

countRow = len(dfCopy)
#split dates up to lowest and highest if there is a dash in the string in the 'date' column
for i in range(countRow):
  dateCheck = dfCopy['Date'][i]
  if '–' or '-' or '—' in dateCheck:
    lowest = extract_values_low(dateCheck)
    highest = extract_values_high(dateCheck)
    # Append to the lists
    lowestDate.append(lowest)
    highestDate.append(highest)
  else:
    lowestDate.append(dateCheck)
    highestDate.append(dateCheck)

# Add the new columns to the DataFrame
dfCopy['Lowest Date'] = lowestDate
dfCopy['Highest Date'] = highestDate

# remove month and empty space within month column and append the value to lowest or highest day
def alterDate(month, date, monthColumn):
  if 'Lowest Month' in monthColumn:
    date = date.replace(month, '')
    date = date.replace(" ", "")
    lowestDay.append(date)
  if 'Highest Month' in monthColumn:
    date = date.replace(month, '')
    date = date.replace(" ", "")
    highestDay.append(date)

#function to check if the specific month is written in the date column as a string and returns a numerical value
def check_months(date, monthColumn):
  # if string found, return specified value
  if 'January' in date:
    value = 1
    alterDate('January', date, monthColumn)
  elif 'February' in date:
    value = 2
    alterDate('February', date, monthColumn)
  elif 'March' in date:
    value = 3
    alterDate('March', date, monthColumn)
  elif 'April' in date:
    value = 4
    alterDate('April', date, monthColumn)
  elif 'May' in date:
    value = 5
    alterDate('May', date, monthColumn)
  elif 'June' in date:
    value = 6
    alterDate('June', date, monthColumn)
  elif 'July' in date:
    value = 7
    alterDate('July', date, monthColumn)
  elif 'August' in date:
    value = 8
    alterDate('August', date, monthColumn)
  elif 'September' in date:
    value = 9
    alterDate('September', date, monthColumn)
  elif 'October' in date:
    value = 10
    alterDate('October', date, monthColumn)
  elif 'November' in date:
    value = 11
    alterDate('November', date, monthColumn)
  elif 'December' in date:
    value = 12
    alterDate('December', date, monthColumn)
  else:
    value = False
    alterDate('', date, monthColumn)
  return value

#assign 'Lowest Date' and 'Highest Date' lists, split them into month and day lists respectively
for i in range(countRow):
  monthCheckLow = dfCopy['Lowest Date'][i]
  monthCheckHigh = dfCopy['Highest Date'][i]
  lowMonthNumber = check_months(monthCheckLow,'Lowest Month')
  highMonthNumber = check_months(monthCheckHigh,'Highest Month')
  if highMonthNumber != False:
    lowestMonth.append(lowMonthNumber)
    highestMonth.append(highMonthNumber)
  else:
    lowestMonth.append(lowMonthNumber)
    highestMonth.append(lowMonthNumber)

# Add the new columns to the DataFrame
dfCopy['Lowest Month'] = lowestMonth
dfCopy['Highest Month'] = highestMonth
dfCopy['Lowest Day'] = lowestDay
dfCopy['Highest Day'] = highestDay

#create a list of numbers or false if they do not exist
has_numeric_low = dfCopy['Lowest Day'].str.contains(r'\d', na=False)
has_numeric_high = dfCopy['Highest Day'].str.contains(r'\d', na=False)
lowestDay = []
highestDay = []

# array to state which months have what days in them
days_in_month = {
  1: 31, 2: 28, 3: 31, 4: 30, 5: 31, 6: 30,
  7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31
}

#checks which month it is and assign amount of days
def checkMonthSize(month):
  return days_in_month.get(month, 0)

# appends value to highest or lowest day list
for i in range(countRow):

  if has_numeric_low[i] == False:
    lowestDay.append('1')
  else:
    lowestDay.append(dfCopy['Lowest Day'][i])

  if has_numeric_high[i] == False:
    monthSize = checkMonthSize(dfCopy['Highest Month'][i])
    highestDay.append(monthSize)
  else:
    highestDay.append(dfCopy['Highest Day'][i])

#assigns list to lowest/highest day in dataframe
dfCopy['Lowest Day'] = lowestDay
dfCopy['Highest Day'] = highestDay

dfCopy.loc[65, 'Highest Day'] = 2
#clean up any extra values in lowest day
dfCopy['Lowest Day'] = dfCopy['Lowest Day'].str.replace(' ', '', regex=True)
dfCopy['Lowest Day'] = dfCopy['Lowest Day'].str.strip()
dfCopy['Lowest Day'] = dfCopy['Lowest Day'].str.replace(r'[^0-9.-]', '', regex=True)

#make lowest and highest day andinteger
dfCopy['Lowest Day'] = pd.to_numeric(dfCopy['Lowest Day'], errors='coerce')
dfCopy['Highest Day'] = pd.to_numeric(dfCopy['Highest Day'], errors='coerce')

#create two new datetime columns from the new day and month columns we have created
dfCopy['Start Date'] = pd.to_datetime(dfCopy[['Year', 'Lowest Month', 'Lowest Day']].rename(columns={'Year': 'year', 'Lowest Month': 'month', 'Lowest Day': 'day'}))
dfCopy['End Date'] = pd.to_datetime(dfCopy[['Year', 'Highest Month', 'Highest Day']].rename(columns={'Year': 'year', 'Highest Month': 'month', 'Highest Day': 'day'}))

#make two new data type columns that indicate range of days the event took
dfCopy['Date Range'] = dfCopy['End Date'] - dfCopy['Start Date']
dfCopy['Range In Days'] = dfCopy['Date Range'].dt.days
```

```python
countRow = len(dfCopy)
countRowWorld = len(dfLandArea)

country_1_Itr = []
country_1_Mass = []

#function to find the land area for specific country specified, else append None
def addLandMass(countrySet, column):
  country_Size = []
  country_Itr = []
  for i in range(countRow):
    if dfCopy[countrySet][i]:
      matches = dfLandArea[dfLandArea['Country'] == dfCopy[countrySet][i]].index.tolist()
      if matches:
        country_Itr.append(matches[0])
      else:
        country_Itr.append(None)
    else:
      country_Itr.append(None)

  for i in range(countRow):
    if country_Itr[i] != None:
      country_Size.append(dfLandArea[column][country_Itr[i]])
    else:
      country_Size.append(None)
  return country_Size

columns_to_clean = ['country 2', 'country 3', 'country 4', 'country 5', 'country 6']

# Clean the columns
for column in columns_to_clean:
    dfCopy[column] = dfCopy[column].astype(str).str.strip()

#calls function addLandMass and assigns it to the dataframe
country_1_Size = addLandMass('country 1', 'Land Area(Km²)')
country_2_Size = addLandMass('country 2', 'Land Area(Km²)')
country_3_Size = addLandMass('country 3', 'Land Area(Km²)')
country_4_Size = addLandMass('country 4', 'Land Area(Km²)')
country_5_Size = addLandMass('country 5', 'Land Area(Km²)')
country_6_Size = addLandMass('country 6', 'Land Area(Km²)')

dfCopy['Size km 1'] = country_1_Size
dfCopy['Size km 2'] = country_2_Size
dfCopy['Size km 3'] = country_3_Size
dfCopy['Size km 4'] = country_4_Size
dfCopy['Size km 5'] = country_5_Size
dfCopy['Size km 6'] = country_6_Size
```
